```python
# 9.Convert all columns to numeric (if possible)
df_numeric = df.apply(pd.to_numeric, errors="coerce")

# Create a DataFrame with min and max values
min_max_df = pd.DataFrame({"Min": df_numeric.min(), "Max": df_numeric.max()})

# Print the DataFrame
print(min_max_df)
```

```python
# 10. Outlier Detection (Boxplots for Numeric Features)
n_cols = 4  # Number of plots per row
valid_cols = [col for col in num_cols if df[col].dropna().nunique() > 0]  # Exclude empty columns
num_features = len(valid_cols)

n_rows = -(-num_features // n_cols)  # Ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(y=df_numeric[col], ax=axes[i])
    axes[i].set_title(f"Boxplot of {col}")

# Hide unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

print("\nEDA Completed!")
```
